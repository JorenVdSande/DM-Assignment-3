```python
# IMPORTS

from preprocessor_anomaly import read_dataset
from anomaly_detection import detect_anomalies_forest, detect_anomalies_lof, detect_anomalies_knn
from plots import plot_elbow_method, plot_silhouette_analysis
import matplotlib.pyplot as plt
import pandas as pd
import os

# set working directory to root
os.chdir("..")
```

```python
token_df, X = read_dataset(use_bert=False)
```

```python
_, forest_scores = detect_anomalies_forest(X, 50)
_, lof_scores = detect_anomalies_lof(X, 50)
knn_scores = detect_anomalies_knn(X, k=10)

# Select algorithm to use
scores = forest_scores
```

```python
# Create inspection CSV and submission data

inspection_df = pd.DataFrame({
    'doc_id': token_df['doc_id'],
    'anomaly_score': scores,
    'token_text': token_df['token_text'],
    'text': token_df['text']
})

inspection_df = inspection_df.sort_values(by='anomaly_score').head(50)

output_df = pd.DataFrame({
    'anomaly': range(1, 51),
    'doc_id': inspection_df['doc_id'].values
})

output_df.to_csv("submission_data/anomalies.csv", index=False)
inspection_df.to_csv("data/anomalies.csv", index=False)
```

```python
# length comparison

token_df['char_count'] = token_df['text'].str.len()
avg_normal_len = token_df['char_count'].mean()
median_normal_len = token_df['char_count'].median()

inspection_df['char_count'] = inspection_df['text'].str.len()
avg_anomaly_len = inspection_df['char_count'].mean()
median_anomaly_len = inspection_df['char_count'].median()

print(f"--- Length Analysis ---")
print(f"Average character length (Normal):  {avg_normal_len:.2f}")
print(f"Average character length (Anomalies): {avg_anomaly_len:.2f}")
print(f"Median character length (Normal):  {median_normal_len:.2f}")
print(f"Median character length (Anomalies): {median_anomaly_len:.2f}")
```
