```python
# IMPORTS

from preprocessor import read_dataset
from clustering import run_clustering, inspect_clusters
from plots import plot_elbow_method, plot_silhouette_analysis
from sklearn.cluster import KMeans, AgglomerativeClustering, SpectralClustering
from sklearn.decomposition import TruncatedSVD
import matplotlib.pyplot as plt
import pandas as pd
import os

# set working directory to root
os.chdir("..")
```


Apply clustering algorithm

```python
# Get the data
token_df, X = read_dataset(use_bert=True)
```

```python
# Configure models
n_clusters = 6

kmeans = KMeans(n_clusters=n_clusters, init='k-means++', random_state=42, n_init=20)

hierarchical = AgglomerativeClustering(n_clusters=n_clusters)

spectral = SpectralClustering(n_clusters=n_clusters, random_state=42, affinity='rbf')

# Select model to use
model = hierarchical
```

```python
# ---------------------
# Run Clustering Method
# ---------------------

labels = run_clustering(X, model)

# Create submission csv
output_df = pd.DataFrame({
    'doc_id': token_df['doc_id'],
    'label': labels
})
output_df.to_csv('data/clusters.csv', index=False)
print("clusters.csv created.")

# Create inspection CSV
inspection_df = pd.DataFrame({
    'doc_id': token_df['doc_id'],
    'label': labels,
    'token_text': token_df['token_text'],
    'text': token_df['text']
})

# Inspect clusters: Look at most frequent tokens
inspect_clusters(inspection_df, n_clusters, 80)

inspection_df.to_csv('data/clusters_and_tokens.csv', index=False)
```
